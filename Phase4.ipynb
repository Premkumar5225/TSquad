```python
# Count null values
null_counts = df.isnull().sum()

# Count zero values
zero_counts = (df == 0).sum()

#DataFrame to store the counts
count_df = pd.DataFrame({'Null Counts': null_counts, 'Zero Counts': zero_counts})

# Plot the graph
count_df.plot(kind='bar', figsize=(10, 6))
plt.title('Null and Zero Value Counts in Each Column')
plt.xlabel('Columns')
plt.ylabel('Counts')
plt.show()
```

```python
#dropping Coloum managerVehicle
df = df.drop('managerVehicle', axis=1)

# Remove null values in distance
df = df.dropna(subset=['distance'])

# Remove zero values in dollars
df = df[df['dollars'] != 0]

# Count null values
null_counts = df.isnull().sum()

# Count zero values
zero_counts = (df == 0).sum()

#DataFrame to store the counts
count_df = pd.DataFrame({'Null Counts': null_counts, 'Zero Counts': zero_counts})

# Plot the graph
count_df.plot(kind='bar', figsize=(10, 6))
plt.title('Null and Zero Value Counts in Each Column')
plt.xlabel('Columns')
plt.ylabel('Counts')
plt.show()
```

We Dropped managerVehicle since it has little impact on our research question and removed all the null count in distance,zero count in dollars for better data model.

.describe(): This method is called on the selected DataFrame and generates summary statistics for each of the selected columns. These statistics include:

count: The number of non-null entries in each column.
mean: The average value of each column.
std: The standard deviation, which measures the spread or dispersion of the values.
min: The minimum value in each column.
25%: The 25th percentile, which is the value below which 25% of the data falls.
50%: The 50th percentile, which is the median or middle value of the data.
75%: The 75th percentile, which is the value below which 75% of the data falls.

```python
#visualize power consumption
plt.figure(figsize=(10,6))
sns.lineplot(df['kwhTotal'])
plt.title('Power Consumption')
plt.xlabel('Event counts')
plt.ylabel('Power (kWh)')
```

```python
# Convert 'startTime' and 'endTime' to datetime objects
df['startTime'] = pd.to_datetime(df['startTime'], unit='s')
df['endTime'] = pd.to_datetime(df['endTime'], unit='s')

# Calculate power consumption per hour
df['powerPerHour'] = df['kwhTotal'] / ((df['endTime'] - df['startTime']).dt.total_seconds() / 3600)

# Group by day of the week and sum the power consumption
daily_power_consumption = df.groupby('weekday')['powerPerHour'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(daily_power_consumption['weekday'], daily_power_consumption['powerPerHour'], color='skyblue')
plt.xlabel('Day of the Week')
plt.ylabel('Power Consumption per Hour (kWh)')
plt.title('Daily Power Consumption')
plt.show()
```

```python
df['startTime'] = pd.to_datetime(df['startTime'], unit='s')
df['endTime'] = pd.to_datetime(df['endTime'], unit='s')

# Calculate wait time duration for each day
df['waitTime'] = (df['endTime'] - df['startTime']).dt.total_seconds() / 3600  # in hours

# Group by weekday and calculate the average wait time for each day
avg_wait_time = df.groupby('weekday')['waitTime'].mean()

# Plotting the average wait time
plt.figure(figsize=(10, 6))
avg_wait_time.plot(kind='bar', color='skyblue')
plt.title('Average Daily Wait Time Duration')
plt.xlabel('Weekday')
plt.ylabel('Average Wait Time (Hours)')
plt.xticks(rotation=45)
plt.show()
```
